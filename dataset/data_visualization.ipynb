```python
from torchaudio.datasets import LIBRISPEECH
```

```python
dataset = LIBRISPEECH("./", download=False)
```

```python
import random
# Get random example
def get_random_set(dataset):
    random_idx = random.randint(0, len(dataset))
    wf, sr, transcript, _, _, _ = dataset[random_idx]
    return {
        "waveform": wf[0],
        "sampling_rate": sr,
        "transcript": transcript.lower()
    }
```

```python
random_set = get_random_set(dataset)
```

```python
import librosa.display
import matplotlib.pyplot as plt

plt.figure()
librosa.display.waveshow(random_set["waveform"].numpy(), sr=random_set["sampling_rate"])
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
